```python
# Python 👉 “OOP Mini Project (Inheritance Based Login System)”
# Parent Class
class User:

    def __init__(self, username, pin):

        self.username = username
        self.pin = pin


    def login(self):

        user = input("Enter Username: ")
        user_pin = input("Enter PIN: ")


        if user == self.username and user_pin == self.pin:

            print("Login Successful")
            return True

        else:
            print("Invalid Username or PIN")
            return False



# Student Class
class Student(User):

    def study(self):
        print(self.username, "is studying")



# Teacher Class
class Teacher(User):

    def teach(self):
        print(self.username, "is teaching")



# Admin Class
class Admin(User):

    def manage_system(self):
        print(self.username, "is managing system")



# Objects
student = Student("muslim", "1111")
teacher = Teacher("rahim", "2222")
admin = Admin("admin", "9999")



# Menu
print("1. Student Login")
print("2. Teacher Login")
print("3. Admin Login")


choice = input("Enter Choice: ")



# Student Login
if choice == "1":

    if student.login():
        student.study()



# Teacher Login
elif choice == "2":

    if teacher.login():
        teacher.teach()



# Admin Login
elif choice == "3":

    if admin.login():
        admin.manage_system()



else:
    print("Invalid Choice")
```
