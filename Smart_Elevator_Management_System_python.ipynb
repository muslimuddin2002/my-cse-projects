```python
# Smart Elevator Management System

class Elevator:

    def __init__(self, elevator_id, total_floors):
        self.elevator_id = elevator_id
        self.current_floor = 1
        self.total_floors = total_floors
        self.door_status = "Closed"
        self.power_status = True
        self.emergency_light = False

    # Weight Check
    def check_weight(self, weight):

        if weight > 1000:

            print("\n🔴 RED ALERT 🔴")
            print("OVERLOAD DETECTED!")
            print(f"Current Weight: {weight} KG")
            print("Maximum Capacity: 1000 KG")

            print("\n🚫 Elevator Locked")
            print("🔔 Alarm Activated")

            self.send_overload_sms(weight)

            return False

        print(f"\n✅ Weight OK ({weight} KG)")
        return True

    # Elevator Move
    def move(self, target_floor):

        if not self.power_status:
            print("\n⚠ Power Failure!")
            print("Elevator Cannot Move.")
            return

        if target_floor < 1 or target_floor > self.total_floors:
            print("Invalid Floor!")
            return

        print(f"\nElevator {self.elevator_id} Moving...")

        while self.current_floor < target_floor:
            self.current_floor += 1
            print(f"⬆ Going Up -> Floor {self.current_floor}")

        while self.current_floor > target_floor:
            self.current_floor -= 1
            print(f"⬇ Going Down -> Floor {self.current_floor}")

        print(f"\n✅ Reached Floor {self.current_floor}")

    # Door Functions
    def open_door(self):
        self.door_status = "Open"
        print("🚪 Door Opened")

    def close_door(self):
        self.door_status = "Closed"
        print("🚪 Door Closed")

    # Power Failure
    def power_failure(self):

        self.power_status = False
        self.emergency_light = True

        print("\n⚠ POWER FAILURE DETECTED!")
        print("💡 Emergency Light ON")

        self.send_power_sms()

    # Restore Power
    def restore_power(self):

        self.power_status = True
        self.emergency_light = False

        print("\n✅ Power Restored")
        print("💡 Emergency Light OFF")

    # SMS for Power Failure
    def send_power_sms(self):

        print("\n📱 SMS SENT TO AUTHORITY")
        print("----------------------")
        print("Lift Power Failure!")
        print(f"Elevator ID: {self.elevator_id}")
        print(f"Current Floor: {self.current_floor}")

    # SMS for Overload
    def send_overload_sms(self, weight):

        print("\n📱 SMS SENT TO AUTHORITY")
        print("----------------------")
        print("OVERLOAD ALERT!")
        print(f"Elevator ID: {self.elevator_id}")
        print(f"Current Floor: {self.current_floor}")
        print(f"Weight: {weight} KG")

    # Emergency Call
    def emergency_call(self):

        print("\n🚨 EMERGENCY CALL ACTIVATED")
        print("Connecting Security Desk...")
        print("📞 Call Connected Successfully")

    # Status
    def status(self):

        print("\n========== ELEVATOR STATUS ==========")
        print("Elevator ID      :", self.elevator_id)
        print("Current Floor    :", self.current_floor)
        print("Door Status      :", self.door_status)
        print("Power Status     :", "ON" if self.power_status else "OFF")
        print("Emergency Light  :", "ON" if self.emergency_light else "OFF")
        print("====================================")


# Child Class (Inheritance)
class VIPElevator(Elevator):

    def move(self, target_floor):
        print("\n⭐ VIP Elevator Activated ⭐")
        super().move(target_floor)


# Objects
normal_lift = Elevator("A1", 10)
vip_lift = VIPElevator("VIP1", 10)


# Main Program
while True:

    print("\n")
    print("===================================")
    print(" SMART ELEVATOR MANAGEMENT SYSTEM ")
    print("===================================")
    print("1. Call Normal Elevator")
    print("2. Call VIP Elevator")
    print("3. Check Elevator Status")
    print("4. Simulate Power Failure")
    print("5. Restore Power")
    print("6. Emergency Call")
    print("7. Exit")

    choice = input("\nEnter Choice: ")

    # Normal Elevator
    if choice == "1":

        floor = int(input("Enter Destination Floor: "))
        weight = float(input("Enter Total Weight (KG): "))

        if normal_lift.check_weight(weight):

            normal_lift.move(floor)
            normal_lift.open_door()
            normal_lift.close_door()

    # VIP Elevator
    elif choice == "2":

        floor = int(input("Enter Destination Floor: "))
        weight = float(input("Enter Total Weight (KG): "))

        if vip_lift.check_weight(weight):

            vip_lift.move(floor)
            vip_lift.open_door()
            vip_lift.close_door()

    # Status
    elif choice == "3":

        print("\n1. Normal Lift")
        print("2. VIP Lift")

        lift_choice = input("Select Lift: ")

        if lift_choice == "1":
            normal_lift.status()

        elif lift_choice == "2":
            vip_lift.status()

        else:
            print("Invalid Choice!")

    # Power Failure
    elif choice == "4":

        print("\n1. Normal Lift")
        print("2. VIP Lift")

        lift_choice = input("Select Lift: ")

        if lift_choice == "1":
            normal_lift.power_failure()

        elif lift_choice == "2":
            vip_lift.power_failure()

    # Restore Power
    elif choice == "5":

        print("\n1. Normal Lift")
        print("2. VIP Lift")

        lift_choice = input("Select Lift: ")

        if lift_choice == "1":
            normal_lift.restore_power()

        elif lift_choice == "2":
            vip_lift.restore_power()

    # Emergency Call
    elif choice == "6":

        print("\n1. Normal Lift")
        print("2. VIP Lift")

        lift_choice = input("Select Lift: ")

        if lift_choice == "1":
            normal_lift.emergency_call()

        elif lift_choice == "2":
            vip_lift.emergency_call()

    # Exit
    elif choice == "7":

        print("\nSystem Closed Successfully.")
        break

    else:
        print("\n❌ Invalid Choice!")
```
